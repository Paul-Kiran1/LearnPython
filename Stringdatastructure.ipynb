**Write a program to access each character of string in forward and backward direction
by using while loop**

```python
 s="Learning python is very good"
 n=len(s)
 i=0
 print("Forward direction")
 while i<n:
     print(s[i],end='')
     i+=1
 print("\nBackward direction :")
 i=-1
 while i>=-n:
     print(s[i],end='')
     i=i-1
```

```python
s = "Learning Python is very easy !!!"

print("Forward direction:")
for ch in s:
    print(ch, end='')

print("\nBackward direction:")
for ch in s[::-1]:    # slicing to reverse
    print(ch, end='')
```
